```python
null_values = []
for col in data.columns:
    if data[col].isna().sum() != 0:
        pct_na = np.round((100 * (data[col].isna().sum())/len(data)), 2)            
        dict1 ={
            'Признаки' : col,
            'NA (count)': data[col].isna().sum(),
            'NA (%)': pct_na
        }
        null_values.append(dict1)
DF1 = pd.DataFrame(null_values, index=None).sort_values(by='NA (count)',ascending=False)
ax = sns.barplot(y='NA (%)', x='Признаки', data=DF1, palette="pastel")
ax.bar_label(ax.containers[0])
ax.set_title('Процент пропущенных значений')
None
```

Так как сейчас проводится только анализ данных, мне кажется нет смысла заполнять пустые значения. Отмечу, как именно можно их заполнить:
- Для признака `Cabin` очень много пропущенных значений, его можно заменить на новый признак, обозначающий наличие или отсутсвие каюты

```python
f, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))
sns.set_color_codes('pastel')

sns.histplot(data['Age'], kde=True, ax=axes[0])
axes[0].axvline(data['Age'].median(), color='red', ls='--', label = 'median', lw=1)
axes[0].text(data['Age'].median()+1,4000,'median = %2.2f'%(data['Age'].median()), color='red')

axes[0].axvline(data['Age'].mean(), color='black', ls='-.', label = 'mean', lw=1)
axes[0].text(data['Age'].mean()-13,4000,'mean = %2.2f'%(data['Age'].mean()))
axes[0].legend()

sns.kdeplot(data.loc[data['Survived'] == 0]['Age'], shade=True, ax=axes[1], label='Not survived')
sns.kdeplot(data.loc[data['Survived'] == 1]['Age'], color='pink', shade=True, ax=axes[1], label='Survived')

plt.legend()
None
```

### Ticket - билет

В этом признаке можно выделить код билета и номер билета. Рассмотрим только код билета
- Нечисловые значения будут отнесены к категории `TicketCode`(только первая буква)
- Поскольку некоторые билеты не имеют нечислового значения или числового значения, это создаст отсутствующее значение. Отсутствующее значение будет классифицировано как `N` для `TicketCode`.

```python
data['TicketCode'] = data['Ticket'].str.extract('(\D+)')
data['TicketCode'] = data['TicketCode'].fillna('N')
data['TicketCode'] = data['TicketCode'].str[0]
```

```python
data_1_df = data.loc[data[target_column] == 1]
df_1 = data_1_df.groupby('TicketCode')['TicketCode'].count()
df = data.groupby('TicketCode')['TicketCode'].count()

fig, axes = plt.subplots(nrows=1, ncols=3,figsize=(14, 4))

sns.countplot(x='TicketCode', data=data, palette="pastel",ax=axes[0],order=df.sort_values(ascending=False).index)
axes[0].set_title('Распределение по типу билета')
axes[0].set(ylim=(0, 98000))
total = float(len(data))
for p in axes[0].patches:
    percentage = '{:.3f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()*0.7
    y = p.get_height()
    axes[0].annotate(percentage, (x, y+1500),xytext=(0, 2),textcoords='offset points',ha='right',rotation=90)



sns.countplot(x='TicketCode', data=data, palette="pastel", hue='Survived', hue_order=[1, 0],ax=axes[1], order=df.sort_values(ascending=False).index)
axes[1].set_ylabel("Survival count")
axes[1].set_title('Сравнение количества выживших по по типу билета')


sns.barplot(x = df.index, y = df_1.values / df.values * 100, label = 'Total',ax=axes[2], palette="pastel", order=df.sort_values(ascending=False).index)
axes[2].set(ylim=(0, 100))
axes[2].set_ylabel("Percentage")

for k in axes[2].patches:
    x = k.get_x() + k.get_width()*0.7
    y = k.get_height()
    percentage = '{:.2f}%'.format(y)
    axes[2].annotate(percentage, (x, y+1),xytext=(0, 3),textcoords='offset points',ha='right',rotation=90)
axes[2].set_title('Процент выживших по типу билета')

plt.tight_layout()
```

*Пассажиры, чей код билета начинается с P имеют больше шанса выжить. Пассажиры с другими кодами имеют в основном меньше шанса выжить, чем пассажиры без кода билета, кроме билета типа L*

```python
feature_cols = data.drop(['Survived'], axis=1).columns
numerical_columns = ['Age', 'Fare']
categorical_columns = data[feature_cols].drop(columns=numerical_columns).columns
pure_cat_cols = data[feature_cols].select_dtypes(exclude=['int64','float64']).columns
```

```python
# label encoding
le = LabelEncoder()
le_data = data.copy()

for col in pure_cat_cols:
    le_data[col] = le.fit_transform(le_data[col])  

corrdata = le_data

## correlation 
corr = corrdata.corr().abs()
mask = np.triu(np.ones_like(corr, dtype=np.bool_))
corr1 = corr.mask(mask)
f, ax = plt.subplots(figsize=(11, 9)) 
sns.heatmap(corr1, mask=mask, vmax=.3, center=0, annot=True,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, fmt="1.2f", cmap="BuPu")
plt.title('Global correlation matrix',fontsize=16, fontstyle='italic')
None
```
