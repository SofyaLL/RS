### Количество пропущенных значений

```python
print('Количество пропущенных значений во всем датасете:', sum(data.isnull().sum()))
print('Процент пропущенных значений во всем датасете:', round(sum(data.isnull().sum())/(len(data)*11)*100, 4))
```

Количество пропущенных значений в каждой колонке

```python
data.isna().sum()
```

```python
null_values = []
for col in data.columns:
    if data[col].isna().sum() != 0:
        pct_na = np.round((100 * (data[col].isna().sum())/len(data)), 2)            
        dict1 ={
            'Признаки' : col,
            'NA (count)': data[col].isna().sum(),
            'NA (%)': pct_na
        }
        null_values.append(dict1)
DF1 = pd.DataFrame(null_values, index=None).sort_values(by='NA (count)',ascending=False)
ax = sns.barplot(y='NA (%)', x='Признаки', data=DF1, palette="pastel")
ax.bar_label(ax.containers[0])
ax.set_title('Процент пропущенных значений')
None
```

## Числовые признаки
К числовым признакам (continuous) относятся
- Age
- Fare

### Age - возраст

```python
f, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))
sns.set_color_codes('pastel')

sns.histplot(data['Age'], kde=True, ax=axes[0])
axes[0].axvline(data['Age'].median(), color='red', ls='--', label = 'median', lw=1)
axes[0].text(data['Age'].median()+1,4000,'median = %2.2f'%(data['Age'].median()), color='red')

axes[0].axvline(data['Age'].mean(), color='black', ls='-.', label = 'mean', lw=1)
axes[0].text(data['Age'].mean()-13,4000,'mean = %2.2f'%(data['Age'].mean()))
axes[0].legend()

sns.kdeplot(data.loc[data['Survived'] == 0]['Age'], shade=True, ax=axes[1], label='Not survived')
sns.kdeplot(data.loc[data['Survived'] == 1]['Age'], color='pink', shade=True, ax=axes[1], label='Survived')

plt.legend()
None
```

*Кажется, что люди в возрасте от 18 до 42 имеют тенденцию умирать с большей вероятностью, чем люди в возрасте от 40 до 70.*

### Fare - стоимость билета

```python
f, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))
sns.set_color_codes('pastel')

sns.histplot(data['Fare'], kde=True, ax=axes[0])
axes[0].axvline(data['Fare'].median(), color='red', ls='--', label = 'median', lw=1)
axes[0].text(data['Fare'].median()+50,4000,'median = %2.2f'%(data['Fare'].median()), color='red')

axes[0].axvline(data['Fare'].mean(), color='black', ls='-.', label = 'mean', lw=1)
axes[0].text(data['Fare'].median()+50,4500,'mean = %2.2f'%(data['Fare'].mean()))
axes[0].legend()

sns.kdeplot(data.loc[data['Survived'] == 0]['Fare'], shade=True, ax=axes[1], label='Not survived')
sns.kdeplot(data.loc[data['Survived'] == 1]['Fare'], color='pink', shade=True, ax=axes[1], label='Survived')

plt.legend()
None
```

*Те, у кого более низкие значения стоимости `fare`, как правило, умирают чаще, чем те, у кого более высокие значения.*

###  Pclass - класс билета

```python
data_1_df = data.loc[data[target_column] == 1]
df_1 = data_1_df.groupby('Pclass')['Pclass'].count()
df = data.groupby('Pclass')['Pclass'].count()

fig, axes = plt.subplots(nrows=1, ncols=3,figsize=(12, 4))

sns.countplot(x='Pclass', data=data, palette="pastel",ax=axes[0])
axes[0].set_title('Распределение по классам')
total = float(len(data))
for p in axes[0].patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()*0.7
    y = p.get_height()
    axes[0].annotate(percentage, (x, y),xytext=(0, 2),textcoords='offset points',ha='right')



sns.countplot(x='Pclass', data=data, palette="pastel", hue='Survived', hue_order=[1, 0],ax=axes[1])
axes[1].set_ylabel("Survival count")
axes[1].set_title('Сравнение количества выживших по классам')


sns.barplot(x = df.index, y = df_1.values / df.values * 100, label = 'Total',ax=axes[2], palette="pastel")
axes[2].set(ylim=(0, 100))
axes[2].set_ylabel("Percentage")

for k in axes[2].patches:
    x = k.get_x() + k.get_width()*0.7
    y = k.get_height()
    percentage = '{:.2f}%'.format(y)
    axes[2].annotate(percentage, (x, y),xytext=(0, 3),textcoords='offset points',ha='right')
axes[2].set_title('Процент выживших по классам')

plt.tight_layout()
```

*Более высокий класс билета соответствует большей вероятности выжить, это может быть результатом приоритета спасательных шлюпок в зависимости от класса билета.*

### Family
Создадим новые признаки: размер и наличие семьи

```python
data['Famsize'] = data['SibSp'] + data['Parch'] + 1

def hasfam(x) : 
    return "With family" if x.Famsize != 1 else "Alone"

data['Hasfam'] = data.apply(hasfam, axis=1)
```

### Cabin - каюта

Только около 30% всех пассажиров имеют каюту. Некоторые каютные номера начинаются с буквы, которая указывается на палубу, на которой находилась каюта. Добавим новые признаки: обозначение палубы и наличие каюты.

```python
data['Deck'] = data['Cabin'].apply(lambda x: 'N' if pd.isnull(x) else x[0])
data['HasCabin'] = data['Deck'].apply(lambda x: 'Cabin' if x!='N' else 'NoCabin')
```

```python
data_1_df = data.loc[data[target_column] == 1]
df_1 = data_1_df.groupby('Deck')['Deck'].count()
df = data.groupby('Deck')['Deck'].count()

fig, axes = plt.subplots(nrows=1, ncols=3,figsize=(13, 4))

sns.countplot(x='Deck', data=data, palette="pastel",ax=axes[0],order=df.sort_values(ascending=False).index)
axes[0].set_title('Распределение по палубам')
axes[0].set(ylim=(0, 90000))
total = float(len(data))
for p in axes[0].patches:
    percentage = '{:.3f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()*0.7
    y = p.get_height()
    axes[0].annotate(percentage, (x, y+1500),xytext=(0, 2),textcoords='offset points',ha='right',rotation=90)



sns.countplot(x='Deck', data=data, palette="pastel", hue='Survived', hue_order=[1, 0],ax=axes[1], order=df.sort_values(ascending=False).index)
axes[1].set_ylabel("Survival count")
axes[1].set_title('Сравнение количества выживших по палубам')


sns.barplot(x = df.index, y = df_1.values / df.values * 100, label = 'Total',ax=axes[2], palette="pastel", order=df.sort_values(ascending=False).index)
axes[2].set(ylim=(0, 100))
axes[2].set_ylabel("Percentage")

for k in axes[2].patches:
    x = k.get_x() + k.get_width()*0.7
    y = k.get_height()
    percentage = '{:.2f}%'.format(y)
    axes[2].annotate(percentage, (x, y+1),xytext=(0, 3),textcoords='offset points',ha='right',rotation=90)
axes[2].set_title('Процент выживших по палубам')

plt.tight_layout()
```

### Ticket - билет

В этом признаке можно выделить код билета и номер билета. Рассмотрим только код билета
- Нечисловые значения будут отнесены к категории `TicketCode`(только первая буква)
- Поскольку некоторые билеты не имеют нечислового значения или числового значения, это создаст отсутствующее значение. Отсутствующее значение будет классифицировано как `N` для `TicketCode`.

```python
data['TicketCode'] = data['Ticket'].str.extract('(\D+)')
# data['TicketCode'] = data['TicketCode'].str.replace('[^\w\s]','')
# data['TicketCode'] = data['TicketCode'].str.replace(' ','')
# data['TicketNumber'] = data['Ticket'].str.extract('(\d+)')
# data['TicketNumber'] = data['TicketNumber'].astype(float)
data['TicketCode'] = data['TicketCode'].fillna('N')
# data['TicketNumber'] = data['TicketNumber'].fillna(0)
data['TicketCode'] = data['TicketCode'].str[0]
```

```python
data_1_df = data.loc[data[target_column] == 1]
df_1 = data_1_df.groupby('TicketCode')['TicketCode'].count()
df = data.groupby('TicketCode')['TicketCode'].count()

fig, axes = plt.subplots(nrows=1, ncols=3,figsize=(14, 4))

sns.countplot(x='TicketCode', data=data, palette="pastel",ax=axes[0],order=df.sort_values(ascending=False).index)
axes[0].set_title('Распределение по типу билета')
axes[0].set(ylim=(0, 98000))
total = float(len(data))
for p in axes[0].patches:
    percentage = '{:.3f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()*0.7
    y = p.get_height()
    axes[0].annotate(percentage, (x, y+1500),xytext=(0, 2),textcoords='offset points',ha='right',rotation=90)



sns.countplot(x='TicketCode', data=data, palette="pastel", hue='Survived', hue_order=[1, 0],ax=axes[1], order=df.sort_values(ascending=False).index)
axes[1].set_ylabel("Survival count")
axes[1].set_title('Сравнение количества выживших по по типу билета')


sns.barplot(x = df.index, y = df_1.values / df.values * 100, label = 'Total',ax=axes[2], palette="pastel", order=df.sort_values(ascending=False).index)
axes[2].set(ylim=(0, 100))
axes[2].set_ylabel("Percentage")

for k in axes[2].patches:
    x = k.get_x() + k.get_width()*0.7
    y = k.get_height()
    percentage = '{:.2f}%'.format(y)
    axes[2].annotate(percentage, (x, y+1),xytext=(0, 3),textcoords='offset points',ha='right',rotation=90)
axes[2].set_title('Процент выживших по типу билета')

plt.tight_layout()
```

*Пассажиры, чей код билета начинается с P имеют больше шанса выжить. Пассажиры с другими кодами имеют в основном меньше шанса выжить, чем пассажиры без кода билета*

### Sex

Сравним пол и стоимость проезда

```python
cols=['Age','Fare']
num_cols = 2
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12, 3))
for index, column in enumerate(cols):
    sns.boxplot(x=column, y='Sex', data=data, palette="pastel",ax=axes[index])
   
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12, 4))
fig.suptitle('Выживаемость',fontsize=16, fontstyle='italic')
for index, column in enumerate(cols):
    sns.boxplot(x=column, y='Sex', data=data, palette="pastel",hue='Survived',ax=axes[index])
    axes[index].legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

- *В среднем среди пассажиров возраст женщин выше, чем у мужчин*
- *Чем старше пассажиры, тем больше вероятность выжить как у мужчин, так и у женщин*   
- *Средняя стоимость проезда, оплачиваемая пассажирами-мужчинами, была ниже, чем у пассажиров-женщин*

```python
cols=['Pclass','HasCabin', 'Deck','TicketCode','Hasfam','Famsize','Embarked']
fig, axes = plt.subplots(nrows=7, ncols=1,figsize=(10, 18))
for index, column in enumerate(cols):
    sns.countplot(x=column, hue='Sex', data=data, palette="pastel",ax=axes[index])
    axes[index].legend(bbox_to_anchor=(-0.18, 0.5), loc='center', borderaxespad=0.)
    axes[index].set(ylim=(0, 40000)) if index == 0 else axes[index].set(ylim=(0, 60000))
    axes[index].set_title(column,fontsize=16, fontstyle='italic')
    for p in axes[index].patches:
        axes[index].annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.2, p.get_height()+2000),rotation=90)


plt.tight_layout()
```

```python
cols_h = ['Pclass', 'HasCabin', 'Deck', 'TicketCode', 'Hasfam', 'Embarked']

fig, axes = plt.subplots(nrows=7, ncols=1,figsize=(10, 18),constrained_layout=True)

for index, column in enumerate(cols):
    sns.heatmap(data.groupby(['Sex', column])['Survived'].agg('mean').unstack()*100, ax=axes[index],
                square=False, annot=True, fmt='.2f', linewidth=2, 
                cbar_kws={"orientation": "horizontal"}, cmap=sns.color_palette("ch:s=-.6,r=.6", as_cmap=True)
               ) 
#     axes[index].set_xlabel(column)
    axes[index].set_title('Sex & %s' %column,fontsize=16)

# plt.tight_layout()
```

- *Мужчины с типом билета P имели больше шанса выжить, чем в среднем*
- *Мужчины с большим количеством членов семьи имели меньше шанса выжить, чем женщины*
